# **Exploratory Data Analysis**

```python
df_fare.shape
```

```python
df_fare.dtypes
```

```python
### Changing Data Type of Pick and Drop Time into Datetime Data Type using to_datetime() function.

df_fare['tpep_pickup_datetime'] = pd.to_datetime(df_fare['tpep_pickup_datetime'])
```

```python
### Changing Data Type of Pick and Drop Time into Datetime Data Type

df_fare['tpep_dropoff_datetime'] = pd.to_datetime(df_fare['tpep_dropoff_datetime'])
```

```python
df_fare.dtypes
```

```python
df_fare['Total_Duration (Minutes)'] = df_fare['tpep_dropoff_datetime'] - df_fare['tpep_pickup_datetime']
```

```python
df_fare.head()
```

```python
df_fare['Total_Duration (Minutes)'] = df_fare['Total_Duration (Minutes)'].dt.total_seconds()/60
```

```python
df_fare.head()
```

```python
df_fare_filtered = df_fare[['passenger_count', 'payment_type', 'trip_distance', 'fare_amount', 'Total_Duration (Minutes)']]
```

```python
df_fare_filtered.head()
```

```python
df_fare_filtered.isnull().sum()
```

```python
df_fare_filtered.dropna(inplace=True)
```

```python
df_fare_filtered.isnull().sum()
```

```python
df_fare_filtered[['passenger_count', 'payment_type',]] = df_fare_filtered[['passenger_count', 'payment_type',]].astype('int64')
```

```python
df_fare_filtered[df_fare_filtered.duplicated()]
```

```python
df_fare_filtered[df_fare_filtered.duplicated()].shape
```

```python
df_fare_filtered.drop_duplicates(inplace=True)
```

```python
df_fare_filtered.shape
```

```python
df_fare_filtered['passenger_count'].value_counts(normalize=True)
```

```python
df_fare_filtered['payment_type'].value_counts(normalize=True)
```

```python
df_fare_filtered = df_fare_filtered[df_fare_filtered['payment_type'] < 3]
df_fare_filtered = df_fare_filtered[(df_fare_filtered['passenger_count'] > 0) & (df_fare_filtered['passenger_count'] < 6)]
```

```python
df_fare_filtered.shape
```

```python
df_fare_filtered['payment_type'].replace([1,2], ['Card', 'Cash'], inplace=True)
```

```python
df_fare_filtered
```

```python
df_fare_filtered[df_fare_filtered['payment_type'] == 'Card']
```

```python
df_fare_filtered.describe()
```

```python
df_fare_filtered = df_fare_filtered[df_fare_filtered['trip_distance'] > 0]
df_fare_filtered = df_fare_filtered[df_fare_filtered['fare_amount'] > 0]
df_fare_filtered = df_fare_filtered[df_fare_filtered['Total_Duration (Minutes)'] > 0]
```

```python
df_fare_filtered
```

```python
df_fare_filtered.describe()
```

```python
plt.hist(df_fare_filtered['trip_distance'])
```

```python
plt.hist(df_fare_filtered['fare_amount'])
```

```python
plt.hist(df_fare_filtered['Total_Duration (Minutes)'])
```

```python
for Col in ['fare_amount', 'passenger_count', 'Total_Duration (Minutes)']:
  q1 = df_fare_filtered[Col].quantile(0.25)
  q3 = df_fare_filtered[Col].quantile(0.75)
  IQR = q3 - q1

  lower_bound = q1 - 1.5 * IQR
  upper_bound = q3 + 1.5 * IQR

  df_fare_filtered =  df_fare_filtered[(df_fare_filtered[Col] >= lower_bound) & (df_fare_filtered[Col] <= upper_bound)]
```

```python
df_fare_filtered
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title('Distribution of Fare Amount')
plt.hist(df_fare_filtered[df_fare_filtered['payment_type'] == 'Card']['fare_amount'], histtype = 'barstacked', bins = 20, edgecolor = 'k', color = '#FA643F', label = 'Card')
plt.hist(df_fare_filtered[df_fare_filtered['payment_type'] == 'Cash']['fare_amount'], histtype = 'barstacked', bins = 20, edgecolor = 'k', color = '#FFBCAB', label = 'Cash')
plt.legend()

plt.subplot(1,2,2)
plt.title('Distribution of Trip Distance')
plt.hist(df_fare_filtered[df_fare_filtered['payment_type'] == 'Card']['trip_distance'], histtype = 'barstacked', bins = 20, edgecolor = 'k', color = '#FA643F', label = 'Card')
plt.hist(df_fare_filtered[df_fare_filtered['payment_type'] == 'Cash']['trip_distance'], histtype = 'barstacked', bins = 20, edgecolor = 'k', color = '#FFBCAB', label = 'Cash')
plt.legend()
plt.show()
```

```python
df_fare_filtered.groupby('payment_type').agg({'fare_amount': ['mean', 'std'], 'trip_distance': ['mean', 'std']})
```

```python
plt.title('Preference of Payment Type: Card or Cash')
plt.pie(df_fare_filtered['payment_type'].value_counts(normalize = True), labels = df_fare_filtered['payment_type'].value_counts().index, startangle = 90, shadow = True, autopct = '%1.1f%%', colors = ['#FA643F', '#FFBCAB'])
plt.legend()
plt.show()
```
